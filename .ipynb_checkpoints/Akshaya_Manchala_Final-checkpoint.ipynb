### Data Augmentation (Emperical Regularization)

```python
import numpy as np
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.experimental.preprocessing.RandomFlip(mode='horizontal'),
    tf.keras.layers.experimental.preprocessing.RandomZoom(height_factor=(-0.2, 0.2), width_factor=(-0.2, 0.2)),
    tf.keras.layers.experimental.preprocessing.RandomRotation(factor=0.2),
    tf.keras.layers.experimental.preprocessing.RandomContrast(factor=0.2),
    tf.keras.layers.experimental.preprocessing.RandomTranslation(height_factor=(-0.2, 0.2), width_factor=(-0.2, 0.2))
])


train_keras_dataset = train_keras_dataset.map(lambda x, y: (data_augmentation(x), y))
```

### ResNet50

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import GlobalAveragePooling2D,Dropout,BatchNormalization
from tensorflow.keras.losses import CategoricalCrossentropy

model = tf.keras.Sequential()


model.add(tf.keras.applications.resnet50.ResNet50(
        include_top=False,
        weights='imagenet',
        input_shape=(256,256,3),
        classes=5))

for layer in model.layers[0].layers:
    layer.trainable = False


model.add(GlobalAveragePooling2D())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Dense(5, activation='softmax'))


model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])

early_stop = EarlyStopping(monitor='val_loss',patience = 10, restore_best_weights=True)


history = model.fit(train_keras_dataset, epochs=50 , validation_data=validation_keras_dataset, callbacks=[early_stop])
```

```python
training_loss = history.history["loss"]
validation_loss = history.history["val_loss"]
model_epochs = len(training_loss)
```

#### Plot the training and validation errors vs. epochs

```python
plt.plot([x for x in range(model_epochs)],training_loss,label="training")
plt.plot([x for x in range(model_epochs)],validation_loss,label="validation")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
def get_results(dataset):
    y_train = dataset.map(lambda x, y: y)
    y_train = np.concatenate([y for y in y_train], axis=0)
    y_train = [np.where(x==1)[0][0] for x in y_train]
    return y_train
```

```python
def get_test_results(dataset,model):
    results = model.predict(dataset)
    results = [np.argmax(x) for x in results]
    return results
```

### Video Classification Using Moving Averages

#### Converting the videos to MP4

```python
import os
import subprocess

# set the directory to convert
directory = r'/Users/akshayamanchala/Desktop/DSML Final Project/Data/Sport Videos'

# find all AVI files in the directory and subdirectories
avi_files = []
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith('.avi'):
            avi_files.append(os.path.join(root, file))

# convert AVI files to MP4
for avi_file in avi_files:
    mp4_file = os.path.splitext(avi_file)[0] + '.mp4'
    subprocess.call(['ffmpeg', '-i', avi_file, '-codec', 'copy', mp4_file])

print('Conversion complete!')
```

```python
from tensorflow.keras.preprocessing.image import img_to_array
import statistics

class_names = {'BaseballPitch':0, 'Basketball':1, 'Fencing':2, 'TennisSwing':3, 'VolleyballSpiking':4}
true_labels = []
predictions1 = []
predictions2 = []
predictions3 = []
avg_predictions1 = []
avg_predictions2 = []
avg_predictions3 = []
num_predictions = 0


video_path =r'/Users/akshayamanchala/Desktop/DSML Final Project/Data/Sport Videos'

for dir_path in os.listdir(video_path):
    if not os.path.basename(dir_path).startswith('.'):
        sub_path = os.path.join(video_path,dir_path)

        for filename in os.listdir(sub_path):
            if filename.endswith(".mp4"):
                true_label = class_names[dir_path]
                true_labels.append(true_label) 
                video = cv2.VideoCapture(os.path.join(sub_path,filename))
                num_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
                probabilities1 = []
                probabilities2 = []
                probabilities3 = []
                stride = max(num_frames // 100, 1)
                count = 0
                for frame_num in range(0, num_frames, stride):
                    if count >100:
                        break
                    # Set the video's current frame position
                    video.set(cv2.CAP_PROP_POS_FRAMES, frame_num)

                    # Read the frame at the current position
                    ret, frame = video.read()
                    if not ret:
                        continue
                    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
                    frame = cv2.resize(frame, (256, 256))
                    frame = img_to_array(frame)
                    frame = np.expand_dims(frame, axis=0)

                    prediction1 = model.predict(frame)
                    prediction2 = enb_model.predict(frame)
                    prediction3 = vgg_model.predict(frame)

                    probabilities1.append(prediction1)
                    probabilities2.append(prediction2)
                    probabilities3.append(prediction3)
                    count += 1

                avg_prediction1 = np.mean(probabilities1, axis=0) 
                avg_predictions1.append(avg_prediction1) 

                avg_prediction2 = np.mean(probabilities2, axis=0) 
                avg_predictions2.append(avg_prediction2) 

                avg_prediction3 = np.mean(probabilities3, axis=0) 
                avg_predictions3.append(avg_prediction3) 
```

#### Resnet Video Classification results:

```python
# Convert the true labels and predictions to numpy arrays
true_labels = np.array(true_labels)
predictions1 = np.argmax(np.concatenate(avg_predictions1, axis=0), axis=1)

# Calculate the confusion matrix, precision, recall, accuracy, and F1 score
conf_mat = confusion_matrix(true_labels, predictions1)
cr = classification_report(true_labels, predictions1)
acc = accuracy_score(true_labels, predictions1)
f1 = f1_score(true_labels, predictions1, average='weighted')
print("Model : ResNet50")
print("Confusion Matrix:\n", conf_mat)
print("Classification Report:\n", cr)
print("Accuracy: ", acc)
print("F1 Score: ", f1)
```

#### EfficientNet Video Classification results:

```python
# Convert the true labels and predictions to numpy arrays
true_labels = np.array(true_labels)
predictions2 = np.argmax(np.concatenate(avg_predictions2, axis=0), axis=1)

# Calculate the confusion matrix, precision, recall, accuracy, and F1 score
conf_mat = confusion_matrix(true_labels, predictions2)
cr = classification_report(true_labels, predictions2)
acc = accuracy_score(true_labels, predictions2)
f1 = f1_score(true_labels, predictions2, average='weighted')
print("Model : EfficientNet")
print("Confusion Matrix:\n", conf_mat)
print("Classification Report:\n", cr)
print("Accuracy: ", acc)
print("F1 Score: ", f1)
```

### Delete the MP4 version of the files

```python
import os
import glob

def delete_files(path, file_extension):
    # Use os.walk to iterate over each folder
    for foldername, subfolders, filenames in os.walk(path):
        for filename in filenames:
            # Check if the file has the right extension
            if filename.endswith(file_extension):
                # Create absolute path to the file
                file_path = os.path.join(foldername, filename)
                # Delete the file
                os.remove(file_path)

# Call the function with the path and the file extension
delete_files('/Users/akshayamanchala/Desktop/DSML Final Project/Data/Sport Videos', '.mp4')
```
